```python
import sys
from datetime import datetime, timedelta
import requests
import fileinput
import os.path
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder

import warnings
warnings.filterwarnings("ignore")

from matplotlib import pyplot
import seaborn as sns

fileFinal  = 'working/final.csv'

df = pd.read_csv(fileFinal,sep=';')
df["data_pesquisa"]= pd.to_datetime(df["data_pesquisa"])
```

```python
# 1062569;CERVEJA ANTARTICA Lata
pds = df[(df['id_produto'] == 1062569)]
```

```python
pyplot.figure(figsize=(22,8))
pyplot.title('Box')
sns.boxplot(x=pds['preco_encontrado'])
pyplot.show()
```

```python
prod_class = pd.crosstab(index=[pds['rede']],columns=pds['bairro'],values=pds['preco_encontrado'],aggfunc='mean',dropna='True')
for (columnName, columnValue) in prod_class.iteritems():
    pyplot.figure(figsize=(22,8))
    pyplot.title('CERVEJA ANTARTICA Lata -> '+columnName)
    sns.boxplot(x=columnValue)
    pyplot.show()
```

```python
for (columnName, columnValue) in prod_class.iteritems():
    pyplot.figure(figsize=(22,8))
    pyplot.title('CERVEJA ANTARTICA Lata -> '+columnName)
    prod_class[columnName].plot.bar()
    pyplot.show()
 
```
