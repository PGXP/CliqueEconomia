```python
import sys
from datetime import datetime, timedelta
import requests
import fileinput
import os.path
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder

import warnings
warnings.filterwarnings("ignore")

from matplotlib import pyplot
import seaborn as sns

if (os.path.exists('img/produto_classificacao') == False):
    os.makedirs('img/produto_classificacao', exist_ok=True)

fileFinal  = 'working/final.csv'

df = pd.read_csv(fileFinal,sep=',')
df["data_pesquisa"]= pd.to_datetime(df["data_pesquisa"])
```

```python
df.groupby('produto_classificacao').size().plot.bar(figsize=(22,8))
```

```python
rs_class_prod = pd.crosstab(index=[df['razao_social']],columns=df['produto_classificacao'],values=df['preco_encontrado'],aggfunc='mean',dropna='True')
for (columnName, columnValue) in rs_class_prod.iteritems():
    pyplot.figure(figsize=(22,8))
    print('Classificação - Razao Social - '+columnName)
    rs_class_prod[columnName].plot.bar()
    pyplot.show()
    pyplot.savefig('/kaggle/working/img/produto_classificacao/graf_Razao_Social_'+columnName+'.png')
```

```python
bairro_class_prod = pd.crosstab(index=df['bairro'],columns=df['produto_classificacao'],values=df['preco_encontrado'],aggfunc='mean',dropna='True')
for (columnName, columnValue) in bairro_class_prod.iteritems():
    pyplot.figure(figsize=(22,8))
    pyplot.title('Classificação - Bairro - '+columnName)
    bairro_class_prod[columnName].plot.bar()
    pyplot.show()
    pyplot.savefig('/kaggle/working/img/produto_classificacao/graf_Bairro_'+columnName+'.png')
```

```python
rede_class_prod = pd.crosstab(index=[df['rede']],columns=df['produto_classificacao'],values=df['preco_encontrado'],aggfunc='mean',dropna='True')
for (columnName, columnValue) in rede_class_prod.iteritems():
    pyplot.figure(figsize=(22,8))
    pyplot.title('Classificação - Rede - '+columnName)
    rede_class_prod[columnName].plot.bar()
    pyplot.show()
    pyplot.savefig('/kaggle/working/img/produto_classificacao/graf_Rede_'+columnName+'.png')
```

```python
filial_class_prod = pd.crosstab(index=[df['filial'],df['bairro']],columns=df['produto_classificacao'],values=df['preco_encontrado'],aggfunc='mean',dropna='True')
for (columnName, columnValue) in filial_class_prod.iteritems():
    pyplot.figure(figsize=(22,8))
    pyplot.title('Classificação - Filial -  '+columnName)
    filial_class_prod[columnName].plot.bar()
    pyplot.show()
    pyplot.savefig('/kaggle/working/img/produto_classificacao/graf_Filial_'+columnName+'.png')
```

```python
prod = pd.crosstab(index=[df['data_pesquisa']],columns=[df['produto_classificacao']],values=df['preco_encontrado'],aggfunc='mean',dropna='False')
for (columnName, columnValue) in prod.iteritems():
    pyplot.figure(figsize=(22,8))
    pyplot.title('Classificação -  '+str(columnName))
    sns.boxplot(x=columnValue)
    pyplot.show()
    pyplot.savefig('/kaggle/working/img/produto_classificacao/boxplot_'+str(columnName)+'.png')
```

```python
prod_class = pd.crosstab(index=[df['data_pesquisa']],columns=df['produto_classificacao'],values=df['preco_encontrado'],aggfunc='mean',dropna='True')
for (columnName, columnValue) in prod_class.iteritems():
    pyplot.figure(figsize=(22,8))
    pyplot.title('Classificação -  '+columnName)
    prod_class[columnName].plot()
    pyplot.show()
    pyplot.savefig('/kaggle/working/img/produto_classificacao/graf_'+columnName+'.png')
```
